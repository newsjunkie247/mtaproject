```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
census = pd.read_csv('https://raw.githubusercontent.com/newsjunkie247/mtaproject/master/all_140_in_36.P1.csv')
```

```python
#getting NYC values
censusnyc = census[(census["COUNTY"] == 47) | (census["COUNTY"] == 61)|(census["COUNTY"] == 5)| (census["COUNTY"] == 81)|(census["COUNTY"] == 85)]
```

```python
censusnyc
```

```python
#calculating population change
censusnyc['popchange'] = censusnyc['POP100'] - censusnyc['POP100.2000']
```

```python
censusnyc.dtypes
```

```python
censusnyc['POP100']  = censusnyc['POP100'].astype(float)
```

```python
censusnyc.dtypes
```

```python
censusnyc['POP100.2000']  = censusnyc['POP100.2000'].astype(float)
```

```python
censusnyc['percent'] = (censusnyc['popchange'])/(censusnyc['POP100.2000'])*100
```

```python
censusnyc
```

```python
censusnyc = censusnyc.replace(r'NaN', 0, regex=True)
```

```python
censusnyc = censusnyc.replace(r'NAN', 0, regex=True)
```

```python
censusnyc = censusnyc.replace(r'inf', 100, regex=True)
```

```python
censusnyc.sort("COUNTY", ascending=True)
```

```python
censusnyc['percent'].describe()
```

```python
#usingKmeans
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5)
km.fit(censusnyc[["percent"]])
```

```python
km.labels_[:10]
```

```python
censusnyc["k-means label"] = km.labels_
```

```python
censusnyc[censusnyc["k-means label"] == 0]["percent"].hist(bins=2)
censusnyc[censusnyc["k-means label"] == 1]["percent"].hist(bins=2)
censusnyc[censusnyc["k-means label"] == 2]["percent"].hist(bins=2)
censusnyc[censusnyc["k-means label"] == 3]["percent"].hist(bins=2)
censusnyc[censusnyc["k-means label"] == 4]["percent"].hist(bins=2)

```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=6)
km.fit(censusnyc[["percent"]])
```

```python
km.labels_[:10]

```

```python
censusnyc["k-means label"] = km.labels_
```

```python
censusnyc[censusnyc["k-means label"] == 0]["percent"].hist(bins=2)
censusnyc[censusnyc["k-means label"] == 1]["percent"].hist(bins=2)
censusnyc[censusnyc["k-means label"] == 2]["percent"].hist(bins=2)
censusnyc[censusnyc["k-means label"] == 3]["percent"].hist(bins=2)
censusnyc[censusnyc["k-means label"] == 4]["percent"].hist(bins=2)
censusnyc[censusnyc["k-means label"] == 5]["percent"].hist(bins=2)
```

```python
censusnyc.groupby("k-means label")["percent"].describe()
```

```python
censusnyc.to_csv("censusnyc.csv")
```

```python
#importing survey answers
answer = pd.read_csv('https://raw.githubusercontent.com/newsjunkie247/mtaproject/master/answer.csv')
```

```python
#focusing on questions about subway service
subway = answer[answer['question_id']==33]
```

```python
subway
```

```python
citizen = pd.read_csv('https://nycdatastables.s3.amazonaws.com/2013-12-03T16:17:33.753Z/citizen.csv')
```

```python
citizen
```

```python
citizen = citizen.rename(columns = {'id':'citizen_id'})
```

```python
citizen
```

```python
survey = pd.merge(subway, citizen, on="citizen_id", how="outer")
```

```python
transition  = survey[:1356]
```

```python
transition = transition[transition['zip_code']>0]
```

```python
transition = transition.replace(r'\N', 0, regex=True)
```

```python
transition = transition[transition['zip_code']>0]
```

```python
transition
```

```python
transition[transition["zip_code"]=="11104"]
```

```python
transition["zip_code"].value_counts()
```

```python
transition1 = transition[["x_location", "y_location", "zip_code"]]
```

```python
#insert column counting zipcodes
```

```python
transition1['count'] = transition1.groupby(['zip_code'])['zip_code'].transform('count')
```

```python
transition1
```

```python
transition1 = transition1.sort('count', ascending=False)
```

```python
#removing zip codes with only one or two responses
transition1 = transition1[transition1['count']>=3]
```

```python
#grouping by zip code and count
tdata = transition1.groupby(['zip_code', 'count']).mean()
```

```python
tdata.sort("x_location", ascending=False)
```

```python
tdata.sort("y_location", ascending=False)
```

```python
tdata["x_location"].describe()
```
